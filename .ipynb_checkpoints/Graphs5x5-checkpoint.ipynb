```python
import matplotlib.pyplot as plt
```

```python
pop_sizes = [50, 100, 200, 500, 1000, 1500, 2000, 2500, 3000, 3500]
fitness_values = [
    2.2442577821537957,
    1.050080235664736,
    2.1931845881005763,
    1.050080235664736,
    1.050080235664736,
    2.2156937953031894,
    2.2442577821537957,
    2.2156937953031894,
    2.1931845881005763,
    2.2156937953031894
]

plt.figure(figsize=(10, 6))
plt.plot(pop_sizes, fitness_values, marker='o', linestyle='-', color='blue')
plt.title('Relação entre POP_SIZE e Fitness na Matriz 5x5 no algoritmo padrão')
plt.xlabel('POP_SIZE')
plt.ylabel('Fitness')
plt.grid(True)
plt.xticks(pop_sizes, rotation=45)
plt.tight_layout()

plt.show()
```
